```python
import pandas as pd
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load data
df = pd.read_excel("Feature_engineering_DS.xlsx")

# Ensure data is sorted by date (important for time series)
df = df.sort_values(by="date").reset_index(drop=True)

# Define features (X) and target (y)
X = df.drop(columns=["L8","date"])
y = df["L8"]

# Time-based split (no randomization)
train_split_index = int(len(df) * 0.7)
val_test_split_index = int(len(df) * 0.85)

X_train, X_val, X_test = X.iloc[:train_split_index], X.iloc[train_split_index:val_test_split_index], X.iloc[val_test_split_index:]
y_train, y_val, y_test = y.iloc[:train_split_index], y.iloc[train_split_index:val_test_split_index], y.iloc[val_test_split_index:]

# scaling
scaler_X = MinMaxScaler()
X_train = scaler_X.fit_transform(X_train)
X_val = scaler_X.transform(X_val)
X_test = scaler_X.transform(X_test)

scaler_y = MinMaxScaler()
y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1)).flatten()
y_val_scaled = scaler_y.transform(y_val.values.reshape(-1, 1)).flatten()
y_test_scaled = scaler_y.transform(y_test.values.reshape(-1, 1)).flatten()

# MODEL (no early stopping)
model = XGBRegressor(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    objective='reg:squarederror'
)

model.fit(X_train, y_train_scaled)   # no callbacks

# Predict
y_pred = model.predict(X_test)

# Metrics
xgb_mse = mean_squared_error(y_test_scaled, y_pred)
xgb_r2 = r2_score(y_test_scaled, y_pred)
xgb_rmse = np.sqrt(xgb_mse)
xgb_mae = mean_absolute_error(y_test_scaled, y_pred)

print(f"MSE: {xgb_mse:.2f}")
print(f"RMSE: {xgb_rmse:.2f}")
print(f"R²: {xgb_r2:.4f}")
print(f"MAE: {xgb_mae:.4f}")
```

```python
from scipy.stats import uniform, randint

param_dist = {
    'n_estimators': randint(100, 1000),  # Number of boosting rounds
    'learning_rate': uniform(0.5,0.1),  # Step size shrinkage
    'max_depth': randint(3, 10),  # Maximum depth of a tree
    'subsample': uniform(0.6, 0.4),  # Fraction of samples for fitting
    'colsample_bytree': uniform(0.6, 0.4),  # Fraction of features for fitting
    'gamma': uniform(0, 0.5), # Minimum loss reduction required to make a further partition
    'reg_alpha': uniform(0, 0.5), # L1 regularization term on weights
    'reg_lambda': uniform(0, 0.5) # L2 regularization term on weights
}
```

## Setup randomized search

Initialize `RandomizedSearchCV` with the XGBoost model, the parameter distribution, the number of iterations, cross-validation strategy, and scoring metric.


**Reasoning**:
Initialize RandomizedSearchCV with the XGBoost model and the defined parameter distribution.



```python
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBRegressor

xgb = XGBRegressor(objective='reg:squarederror', random_state=42)

random_search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_dist,
    n_iter=10,
    cv=5,
    scoring='neg_mean_squared_error',
    random_state=42,
    n_jobs=-1
)
```
